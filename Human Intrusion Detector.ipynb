#### An AI and computer vision enabled program used to detect human intrusion in hazardous or restricted areas

*This project will open a new window using live camera feed.*

```python
import cv2
import numpy as np
import time

vid = cv2.VideoCapture(0)   #to capture CCTV feed

# Check if camera opened successfully
if (vid.isOpened()== False):
  print("Error opening video stream or file")
ret, frame= vid.read()

faceCascade1 = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
faceCascade2 = cv2.CascadeClassifier('haarcascade_profileface.xml')
height,width,layers=frame.shape
FPS = 10

while(vid.isOpened()):
  tim_fps0 = time.perf_counter()
  ret, frame= vid.read()
  frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
  faces1 = faceCascade1.detectMultiScale(
      frame_gray,
      scaleFactor=1.1,
      minNeighbors=5,
      minSize=(70, 70),
      flags=cv2.CASCADE_SCALE_IMAGE)
  faces2 = faceCascade2.detectMultiScale(
      frame_gray,
      scaleFactor=1.1,
      minNeighbors=5,
      minSize=(70, 70),
      flags=cv2.CASCADE_SCALE_IMAGE)

  # Draw a rectangle around the faces/upper bodies
  for (x, y, w, h) in faces1:
      cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
  for (x, y, w, h) in faces2:
      cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
  txt = "People: :"+str(len(faces1)+len(faces2))
  cv2.putText(frame, txt, (50,50), cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),3)

  cv2.imshow('Video', frame)

  if cv2.waitKey(1) == ord('q'):
    break
  while (time.perf_counter()-tim_fps0 < 1./FPS):
    pass

vid.release()
cv2.destroyAllWindows()
```
